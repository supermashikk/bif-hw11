```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar11/breast_cancer_1000_genes.tsv', sep = '\t', index_col = 0)

df['Max'] = df.max(axis = 1)
df['Min'] = df.min(axis = 1)

df['Variance'] = df.var(axis = 1)
df['Q1'] = df.quantile(q = 0.25, axis = 1)
df['Q3'] = df.quantile(q = 0.75, axis = 1)

df['Interquartile range'] = df['Q3'] - df['Q1']
df['Max - Min'] = df['Max'] - df['Min']

df1 = df.sort_values(by = ['Interquartile range'])
df1 = df1.iloc[0:10,]

df2 = df.sort_values(by = ['Max - Min'])
df2 = df2.iloc[0:10,]

df3 = df.sort_values(by = ['Variance'])
df3 = df3.iloc[0:10,]

plt.tight_layout()


df1 = df1.drop(columns = ['Max', 'Min', 'Variance', 'Q1', 'Q3', 'Interquartile range', 'Max - Min'])
df1 = df1.T
df1 = df1.melt(var_name = 'Genes', value_name = 'Expression')

a = sns.boxplot (x = 'Genes', y = 'Expression', data = df1, palette = 'rocket')
a.set_title('IQR')
a.set_xticklabels(a.get_xticklabels(),rotation=90)
```

```python
df2 = df2.drop(columns = ['Max', 'Min', 'Variance', 'Q1', 'Q3', 'Interquartile range', 'Max - Min'])
df2 = df2.T
df2 = df2.melt(var_name = 'Genes', value_name = 'Expression')
plt.tight_layout()

b = sns.boxplot (x = 'Genes', y = 'Expression', data = df2, palette = 'magma')
b.set_title('Max - Min')
b.set_xticklabels(b.get_xticklabels(),rotation=90)
```

```python
df3 = df3.drop(columns = ['Max', 'Min', 'Variance', 'Q1', 'Q3', 'Interquartile range', 'Max - Min'])
df3 = df3.T
df3 = df3.melt(var_name = 'Genes', value_name = 'Expression')
plt.tight_layout()

c = sns.boxplot (x = 'Genes', y = 'Expression', data = df3, palette = 'crest')
c.set_title('Variance')
c.set_xticklabels(c.get_xticklabels(),rotation=90)
```
