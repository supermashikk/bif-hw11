```python
from scipy.stats import poisson
import pandas as pd
import numpy as np

rv = poisson(17)

df = pd.DataFrame()
for i in range (0,15):
    df[i] = rv.rvs(size = 100)
    
df.loc['mean'] = df.mean()
df.loc['s^2'] = df.std()**2
df.loc['mean - lambda'] = np.abs(df.loc['mean'] - 17)
df.loc['s^2 - lambda'] = np.abs(df.loc['s^2'] - 17)
    
print(df)
```
